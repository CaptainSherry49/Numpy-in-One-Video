```python
x = np.array([[1,2,3],[4,5,6],[7,1,0]])
x
```

```python
x.sum(axis=0) # row wise
```

```python
x.sum(axis=1) # Column wise
```

```python
x = x.T # This will transpose the array mean column into row and row into column
```

```python
x
```

```python
for i in x.flat:
    print(i)
```

```python
x.ndim
# This will print the dimension of the array
```
